# Verification Results Visualization
Load CSVs from `results/` and plot AUC / EER heatmaps per age-bin pair.

Run this notebook from the **project root**.

```python
import os
import ast
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

results_dir = 'results'
datasets = ['fgnet', 'cacd']
models   = ['facenet', 'arcface', 'vggface', 'openface']

for ds in datasets:
    for m in models:
        path = os.path.join(results_dir, f'results_agebins_{ds}_{m}.csv')
        if not os.path.exists(path):
            continue
        df = pd.read_csv(path)
        df['bin_pair'] = df['bin_pair'].apply(ast.literal_eval)
        df[['probe_bin', 'gallery_bin']] = pd.DataFrame(df['bin_pair'].tolist(), index=df.index)
        pivot_auc = df.pivot(index='probe_bin', columns='gallery_bin', values='auc')
        plt.figure(figsize=(8, 6))
        sns.heatmap(pivot_auc, annot=True, fmt='.3f', cmap='YlGnBu')
        plt.title(f'AUC Heatmap - {m.upper()} on {ds.upper()}')
        plt.tight_layout()
        plt.show()
```
