# Identification Results Visualization
Load CSVs from `results/` and plot Rank-1 / Rank-5 / Rank-10 bar charts.

Run this notebook from the **project root**.

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

results_dir = 'results'
datasets = ['fgnet', 'cacd']
models   = ['facenet', 'arcface', 'vggface', 'openface']

records = []
for ds in datasets:
    for m in models:
        path = os.path.join(results_dir, f'results_identification_{ds}_{m}.csv')
        if os.path.exists(path):
            df = pd.read_csv(path)
            df['dataset'] = ds
            df['model']   = m
            records.append(df)

all_df = pd.concat(records, ignore_index=True)
print(all_df.head())
```
